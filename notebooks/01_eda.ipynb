## 2. Length Distributions
A histogram showing the word/token counts of the articles.

```python
# Calculate word counts for the articles
df['word_count'] = df['input_text'].apply(lambda x: len(str(x).split()))

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='word_count', hue='label', bins=50, kde=True)
plt.title('Length Distributions (Word/Token Counts)')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```

## 3. Top Tokens (TF-IDF)
Calculating the top-50 TF-IDF words for each class.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

def get_top_tfidf_words(texts, n_top_words=50):
    tfidf = TfidfVectorizer(stop_words='english', max_features=10000)
    tfidf_matrix = tfidf.fit_transform(texts)
    
    sum_tfidf = tfidf_matrix.sum(axis=0)
    words_freq = [(word, sum_tfidf[0, idx]) for word, idx in tfidf.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)
    
    return pd.DataFrame(words_freq[:n_top_words], columns=['Word', 'TF-IDF'])

# Top 50 tokens for each class
# Replace 0 and 1 with actual label values if different
print("Calculating top tokens for class 0 (Fake News)...")
df_class_0 = df[df['label'] == 0]
top_words_0 = get_top_tfidf_words(df_class_0['input_text'])

print("\nCalculating top tokens for class 1 (Real News)...")
df_class_1 = df[df['label'] == 1]
top_words_1 = get_top_tfidf_words(df_class_1['input_text'])

print("\nTop 50 TF-IDF Words - Class 0:")
display(top_words_0)

print("\nTop 50 TF-IDF Words - Class 1:")
display(top_words_1)
```
